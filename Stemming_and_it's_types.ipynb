**Stemming**

Stemming is the process of reducing a word to its word stem that affixes to suffixes and prefixes or to the roots of words known as lemma. Stemming is important in natural language understanding (NLU) and Natural language processing(NLP).

```python
## Classification problem
## Comments of product is a positive review or negative review
## Reviews ---> eating, eat, eaten [going,gone,goes] ---> go

words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

**PorterStemmer**

```python
from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
  print(word+"--->"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
stemming.stem('sitting')  #it works properly for good number of words,but has issues with more few words as well
```

**RegexpStemmer class**

NLTK has RegexpStemmer class with the help of which we can easily implement Regular Expression Stemmer algorithms. It basically takes a single regular expression and removes any prefix or suffix that matches the expression. Let us see an example

```python
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating')   #only it removes the reg in the last
```

**Snowball stemmer**  (better accuracy than porter stemmer)

```python
from nltk.stem import SnowballStemmer
```

```python
snowballsstremmer = SnowballStemmer('english')
```

```python
for word in words:
  print(word+"--->"+snowballsstremmer.stem(word))
```

```python
stemming.stem('fairly'),stemming.stem('sportingly')
```

```python
snowballsstremmer.stem('fairly'),snowballsstremmer.stem('sportingly')
```
