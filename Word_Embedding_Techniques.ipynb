```python
import tensorflow as tf
print("Tensorflow version:", tf.__version__)
print("GPU is", "available" if tf.config.list_physical_devices("GPU") else "not available")
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
sent
```

```python
###vocubulary size
voc_size=500
```

**One hot representation**

```python
onehot_repr=[one_hot(words,voc_size)for words in sent] ##These basically works and gives as index numbers
print(onehot_repr)
```

**Word Embedding Representation**

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
## pre padding
sent_length=8
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
## 10 feature dimesnions
dim=10
```

```python
model=Sequential()
model.add(Embedding(input_dim = voc_size,output_dim = 10,input_shape = (sent_length,)))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
embedded_docs[0]
```

```python
model.predict(np.expand_dims(embedded_docs[0],axis=0))
```

```python
model.predict(embedded_docs)
```
